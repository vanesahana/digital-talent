```python
# Randomisasi Nilai
import random
nilaiTugas=[]
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))
print(nilaiTugas)
```

```python
#Menampilkan Angka Ganjil
for n in nilaiTugas:
    if n%2==1:
        print("Berikut adalah angka ganjil dari 250 data=")
        print(n)
```

```python
#Menampilkan n>70
for n in nilaiTugas:
    if n>70:
         print(n,"terkategorisasi >70")
```

```python
#Menampilkan Rata Rata dan Standar Deviasi
for i in range(len(nilaiTugas)):
    nilaiTugas[i]=int(nilaiTugas[i])
jumlah= 0
for i in range(len(nilaiTugas)):
    jumlah+=nilaiTugas[i]    
rata=jumlah/len(nilaiTugas)
print("Rata-rata=")
print(rata)
sigma=0
for i in range(len(nilaiTugas)):
    hitung=(nilaiTugas[i]-rata)**2
    sigma+=hitung
pembagianN=sigma/(len(nilaiTugas)-1)
standardeviasi=pembagianN**0.5
print("Standar Deviasi=")
print(standardeviasi)
```

```python
#Menampilkan Nilai Terbesar dalam List
terbesar=0
for a in nilaiTugas:
        if a>terbesar:
            terbesar=a
print("Nilai Terbesar adalah=") 
print(terbesar)
```
